### Exercise 1: Line Plot Customization

```python
# TASK: Create a line plot with x values ranging from 0 to 10 and y values as the square of x.
# Customize the plot by adding a title, labels for both axes, and a grid.
```

```python
import matplotlib.pyplot as plt
import numpy as np

#Data
x=np.linspace(0,10,100) #100 points from 0 to 10
y=x**2  # y values as the square of x

#Create the plot
plt.plot(x,y)

#Customize the plot
plt.title("square of x (y=x^2)")
#Title of the plot
plt.xlabel("x values") #x-axis label
plt.ylabel("y values (x^2)") #y-axis label
plt.grid(True)

#Display the plot
plt.show()
```
